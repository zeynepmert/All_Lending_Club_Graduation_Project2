```python
import pandas as pd
print("Risk tablosu okunuyor ve temizleniyor...")
df_risk = pd.read_csv('/Users/feyzacelik/Documents/GitHub/All_Lending_Club_Graduation_Project2/delinquency_risk.csv', low_memory=False)

# ZAMAN SÜTUNLARI -> "Never"
# "mths_since..." ile başlayanlar. Boşsa "Hiç olmadı" demektir.
time_cols = [
    'mths_since_last_delinq',
    'mths_since_last_record',
    'mths_since_last_major_derog',
    'mths_since_recent_bc',
    'mths_since_recent_bc_dlq',
    'mths_since_recent_inq',
    'mths_since_recent_revol_delinq'
]

# Bu sütunların veri tipini 'object' (metin) yapıyoruz ki "Never" yazabilelim
# Yoksa sayı sütununa yazı yazmaya çalışınca hata verebilir
for col in time_cols:
    if col in df_risk.columns:
        df_risk[col] = df_risk[col].fillna("Never")


# SAYAÇ SÜTUNLARI -> 0
# ID ve Zaman sütunları dışındakileri alıyoruz
count_cols = [
    'delinq_2yrs',
    'chargeoff_within_12_mths',
    'collections_12_mths_ex_med',
    'tax_liens',
    'num_tl_120dpd_2m',
    'num_tl_30dpd',
    'num_tl_90g_dpd_24m',
    'num_tl_op_past_12m',
    'num_accts_ever_120_pd'
]

existing_counts = [c for c in count_cols if c in df_risk.columns]
df_risk[existing_counts] = df_risk[existing_counts].fillna(0)

df_risk.to_csv('delinquency_risk_cleaned.csv', index=False)
```
